```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.distance import geodesic

# Load the dataset
data = pd.read_csv(r"C:\Users\yashm\Downloads\archive (3)\uber.csv")

# Data Cleaning and Preparation (Optional)
# ... (e.g., handling missing values, outliers, etc.)

# Exploratory Data Analysis (EDA)
# 1. Basic Statistics
print(data.describe())

# 2. Data Visualization
#   - Histogram of Fare Amounts
plt.figure(figsize=(10, 6))
sns.histplot(data['fare_amount'], bins=30, kde=True)
plt.title('Distribution of Fare Amounts')
plt.xlabel('Fare Amount')
plt.ylabel('Frequency')
plt.show()

#   - Box Plot of Fare Amounts by Passenger Count
plt.figure(figsize=(10, 6))
sns.boxplot(x='passenger_count', y='fare_amount', data=data)
plt.title('Fare Amount by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Fare Amount')
plt.show()

#   - Scatter Plot of Distance vs. Fare Amount
def haversine_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).km

data['distance'] = data.apply(lambda row: haversine_distance(row['pickup_latitude'], row['pickup_longitude'], row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='distance', y='fare_amount', data=data)
plt.title('Distance vs. Fare Amount')
plt.xlabel('Distance (km)')
plt.ylabel('Fare Amount')
plt.show()

# 3. Correlation Analysis
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

# 4. Feature Engineering and Model Building (Optional)
#   - Create features like time of day, day of week, etc.
#   - Train a machine learning model to predict fare amounts.

# 5. Conclusion and Insights
#   - Summarize key findings from the analysis.
#   - Discuss potential implications for Uber's operations and strategies.
```
